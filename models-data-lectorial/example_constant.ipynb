```python
import numpy as np
```

```python
theta_true = 2
sigma = 1
e = np.random.normal(0,sigma,size=5)
y_obs = theta_true + e
```

```python
plt.plot(y_obs,'o')
plt.hlines(y=np.mean(y_obs),xmin=0,xmax=4)
plt.ylabel('y')
plt.xlabel('sample number')
plt.show()
```

```python

def likelihood(theta):
    return (1/(2*np.pi*sigma**2))*np.exp(-(1/(2*sigma**2))*np.sum((y_obs-theta)**2))
```

```python
theta_grid = np.linspace(0,4,500)
likes = np.array([likelihood(theta_i) for theta_i in theta_grid])

plt.plot(theta_grid,likes/np.max(likes))
plt.ylabel(r'$\mathcal{L}(\theta; y)$',fontsize=16)
plt.xlabel(r'$\theta$',fontsize=16)
plt.show()
```

```python
from scipy.optimize import minimize
theta_initial_guess = 0
res = minimize(lambda theta:-np.log(likelihood(theta)), x0=theta_initial_guess,tol=1e-3)

print(res.x[0])
print(theta_grid[np.argmax(likes)])
print(np.mean(y_obs))
```

```python
theta_grid = np.linspace(0,4,500)
likes = np.array([likelihood(theta_i) for theta_i in theta_grid])

plt.plot(theta_grid,likes/np.max(likes))
plt.ylabel(r'$\mathcal{L}(\theta; y)$',fontsize=16)
plt.xlabel(r'$\theta$',fontsize=16)
plt.savefig('like.pdf')
plt.show()
```

```python
theta_grid = np.linspace(0,4,500)
likes = np.array([likelihood(theta_i) for theta_i in theta_grid])

plt.plot(theta_grid,likes/np.max(likes))
plt.hlines(y=0.15,xmin=0,xmax=5,linestyles='--')
plt.vlines(x=theta_grid[np.argmax(likes)],ymin=0,ymax=1,linestyles='--')
plt.vlines(x=theta_true,ymin=0,ymax=1,linestyles='-')
plt.ylabel(r'$\mathcal{L}(\theta; y)$',fontsize=16)
plt.xlabel(r'$\theta$',fontsize=16)
plt.savefig('like-est.pdf')
plt.show()
```

```python
theta_grid = np.linspace(0,4,500)
likes = np.array([likelihood(theta_i) for theta_i in theta_grid])

plt.plot(theta_grid,-np.log(likes/np.max(likes)))
plt.hlines(y=-np.log(0.15),xmin=0,xmax=5,linestyles='--')
plt.vlines(x=theta_grid[np.argmax(likes)],ymin=0,ymax=25,linestyles='--')
plt.vlines(x=theta_true,ymin=0,ymax=25,linestyles='-')
plt.ylabel(r'$-l(\theta; y)$',fontsize=16)
plt.xlabel(r'$\theta$',fontsize=16)
plt.savefig('loglike-est.pdf')
plt.show()
```
